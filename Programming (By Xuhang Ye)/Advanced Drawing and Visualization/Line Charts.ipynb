```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import warnings

warnings.filterwarnings('ignore')

col1 = np.linspace(0, 10, 1001) # 1001 evenly spaced points between 0 and 10
col2 = np.sin(col1) # sine of col1
df = pd.DataFrame({'col1': col1, 'col2': col2}) # create a dataframe with col1 and col2
print(df.head(10))
```

```python
sns.set_theme(style = "darkgrid")
sns.lineplot(x = 'col1', y = 'col2', data = df) 

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl

employment = pd.read_excel('unemployment.xlsx')
print(employment.head(10))
```

```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA')
sns.set_theme(style = "darkgrid")
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender')
plt.show()
```

### `sns.lineplot`

1. **hue**: 'Gender'****

2. **style**: `hue`****
   
3. **markers**: `markers = True`

4. **dashes**: `dashes``False`



```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA')
sns.set_theme(style = "darkgrid")
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', markers = True, dashes = False)
plt.show()
```

### `sns.lineplot`

1. **err_style**: ****`err_style = 'bars'` 

2. **ci**: Confidence Interval`ci = 70` 70%70%



```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA')
sns.set_theme(style = "darkgrid")
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', err_style = 'bars', ci = 70)
plt.show()
```

### `sns.lineplot`

1. **palette**: `palette = 'CMRmap'`'CMRmap'

2. **markers**: `markers = ['o', '>']`'o' '>' 



### `sns.set()`

1. **xtick.labelsize**: x`'xtick.labelsize': 17`x17x

2. **ytick.labelsize**: `xtick.labelsize`y`'ytick.labelsize': 10`y10y

3. **axes.labelsize**: xy`'axes.labelsize': 15`15

4. **axes.grid**: `'axes.grid': False`



### `plt.title()`

1. **'Employment in USA'**: 

2. **fontsize = 20**: 20

3. **fontweight = 'bold'**: `'bold'`

4. **color = 'blue'**: 

5. **fontfamily = 'serif'**: `'serif'`

6. **loc = 'left'**: `'left'`

7. **pad = 20**: 2020



```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA', fontsize = 20, fontweight = 'bold', color = 'blue', fontfamily = 'serif', loc = 'left', pad = 20)
sns.set_theme(style = "darkgrid")
sns.set(rc = {'xtick.labelsize': 17, 'ytick.labelsize': 10, 'axes.labelsize': 15, 'axes.grid': False})
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', dashes = False, palette = 'CMRmap', markers = ['o', '>'])
plt.show()
```

### `gcf()`

`gcf``get_current_figure`FigureMatplotlibFigureFigure

`plt.gcf()``plt.gcf().text()`

```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA', fontsize = 25, fontweight = 'bold', color = 'blue', fontfamily = 'serif', pad = 20)
sns.set_theme(style = "darkgrid")
plt.gcf().text(.2, .84, 'GENDER', fontsize = 20, fontweight = 'bold', color = 'Black', fontfamily = 'serif')
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', dashes = False, palette = 'CMRmap', markers = ['o', '>'])
plt.show()
```
