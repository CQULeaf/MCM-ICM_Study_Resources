```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl

employment = pd.read_excel('unemployment.xlsx')
print(employment.head(10))
```

```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA')
sns.set_theme(style = "darkgrid")
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender')
plt.show()
```

### 关于`sns.lineplot`其中参数的解释

1. **hue**: 这个参数根据某个分类变量（这里是'Gender'）为图表中的不同类别分配不同的颜色。例如，如果有男性和女性两类，每一类都会用不同**颜色**的线表示。

2. **style**: 类似于`hue`，但它改变的是线条的**样式**而不是颜色。在此例中，男性和女性的数据将通过不同的线条样式（如实线、点线等）来区分。
   
3. **markers**: 通常在折线图中，只有线条可见。启用`markers = True`后，每个数据点会被一个小标记符号（如圆点、方块等）标出，使得数据点更为明显。

4. **dashes**: 正常情况下，不同类别的线可能会以实线和虚线等不同方式呈现。将`dashes`设置为`False`意味着图表中的所有线条都将显示为实线，而不考虑其类别。

这些参数共同作用，使得通过颜色、线条样式、数据点标记等多种方式在图表中清晰地展示和区分数据。

```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA')
sns.set_theme(style = "darkgrid")
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', markers = True, dashes = False)
plt.show()
```

### 关于`sns.lineplot`其中参数的解释

1. **err_style**: 这个参数定义了**误差条**的样式。`err_style = 'bars'` 表示使用误差条来表示数据的不确定性或变异性。误差条通常显示在数据点的上下方，提供了关于数据集中变异性的视觉线索。

2. **ci**: 这个参数表示置信区间（Confidence Interval）的百分比。`ci = 70` 意味着图表会显示每个点的70%置信区间。置信区间是一种统计范围估计，表示对于给定的数据点，我们相信其真实值有70%的可能性落在这个区间内，这通常用于表示数据的可靠性和稳定性。

这些参数帮助在图表中展示与数据相关的统计信息，如数据的可靠性和误差范围，从而使数据解释更为完整和准确。

```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA')
sns.set_theme(style = "darkgrid")
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', err_style = 'bars', ci = 70)
plt.show()
```

### 关于`sns.lineplot`中部分参数的解释

1. **palette**: 这个参数用于选择图表中用于表示不同类别的颜色方案。`palette = 'CMRmap'`指定了一个特定的颜色映射方案，名为'CMRmap'。这个方案定义了一系列颜色，用于区分图表中的不同数据类别。颜色映射方案可以影响图表的整体外观和易读性，不同的方案可能适合不同类型的数据和可视化需求。

2. **markers**: 此参数定义了在图表中表示数据点的标记符号的类型。`markers = ['o', '>']`表示使用两种不同的标记符号来表示数据点。这里，'o' 代表圆形标记，'>' 代表三角形标记。当有多个类别的数据时，不同的标记符号可以帮助区分这些类别，使图表更加清晰易懂。

这两个参数都有助于增强图表的视觉表现，使得图表中的不同数据类别更容易区分和理解。

### 关于`sns.set()`函数的解释

1. **xtick.labelsize**: 这个参数设置x轴刻度标签的字体大小。`'xtick.labelsize': 17`意味着x轴上的刻度标签将使用大小为17的字体。这有助于使x轴的标签更加清晰和易读。

2. **ytick.labelsize**: 类似于`xtick.labelsize`，但这个参数是为y轴的刻度标签设置字体大小。`'ytick.labelsize': 10`表示y轴的刻度标签将使用大小为10的字体。这可以帮助平衡图表的整体外观，确保y轴标签既清晰又不过于突出。

3. **axes.labelsize**: 这个参数设置轴标题（即x轴和y轴旁边通常显示的文字）的字体大小。`'axes.labelsize': 15`意味着轴标题将使用大小为15的字体。轴标题通常用于说明每个轴代表的是什么数据，因此适当的字体大小有助于读者快速理解图表内容。

4. **axes.grid**: 这个参数控制是否显示图表中的网格线。`'axes.grid': False`表示不显示网格线。网格线可以帮助读者更精确地估计数据点的位置，但在某些情况下，不显示网格线可以使图表看起来更简洁。

这些参数共同作用，使图表的外观和可读性更加符合特定的展示需求和美学偏好。

### 关于`plt.title()`函数的解释

1. **'Employment in USA'**: 这是图表的标题文本。它会显示在图表的顶部，用于描述图表的主题或所展示的数据内容。

2. **fontsize = 20**: 这个参数设置标题的字体大小为20。字体大小的选择会影响标题的可见性和突出程度。

3. **fontweight = 'bold'**: 这个参数指定标题字体的粗细。`'bold'`意味着标题将以粗体显示，这有助于使标题更加醒目。

4. **color = 'blue'**: 此参数设置标题的颜色。这里使用的是蓝色，颜色的选择可以根据图表的整体配色方案或个人喜好来定。

5. **fontfamily = 'serif'**: 这个参数决定了使用哪种字体系列。`'serif'`是一种有衬线的字体风格，常用于正式或传统的文本。

6. **loc = 'left'**: 通常标题默认居中，但这个参数可以改变标题的对齐方式。`'left'`表示标题将靠左对齐。

7. **pad = 20**: 这个参数设置标题与图表上边缘的距离，单位是像素。这里设置为20，意味着标题和图表顶部将有20像素的间距。

这些参数共同作用，定义了图表标题的外观和位置，使其既符合数据的表达需求，又与整个图表的美学设计协调一致。

```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA', fontsize = 20, fontweight = 'bold', color = 'blue', fontfamily = 'serif', loc = 'left', pad = 20)
sns.set_theme(style = "darkgrid")
sns.set(rc = {'xtick.labelsize': 17, 'ytick.labelsize': 10, 'axes.labelsize': 15, 'axes.grid': False})
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', dashes = False, palette = 'CMRmap', markers = ['o', '>'])
plt.show()
```

### 关于`text`函数，这是一种在图表上任意位置添加文本的方法。

- `plt.gcf().text(.2, .84, 'GENDER')`: 这个命令是在当前的图形（由`plt.gcf()`获取）上添加文本。`.2` 和 `.84` 是文本位置的坐标，这里是相对于图形窗口的比例坐标。`.2` 表示文本在水平方向上位于图形宽度的20%处，而`.84` 表示在垂直方向上位于图形高度的84%处。'GENDER'是要添加的文本。

这样，文本“GENDER”会被放置在图表的指定位置上，这通常用于添加额外的注释或说明，这个功能在标准的图表标签或标题无法满足特定显示需求时特别有用。

### 关于`gcf()`函数

`gcf`是`get_current_figure`的缩写。这个函数的作用是获取当前的图形（Figure）。在Matplotlib中，图形（Figure）是一个窗口或绘图区域，它包含了生成图表的所有元素。简单来说，当你创建一个或多个图表时，每个图表都被绘制在一个图形（Figure）内。

使用`plt.gcf()`通常出现在需要对当前活动的图形进行操作的情况下，比如调整其大小、添加文本或者保存图形。`plt.gcf().text()`是在当前活动的图形上添加文本，这意味着可以在图形的任何位置放置文本，而不仅限于绘图区域。

```python
plt.figure(figsize = (14, 7))
plt.title('Employment in USA', fontsize = 25, fontweight = 'bold', color = 'blue', fontfamily = 'serif', pad = 20)
sns.set_theme(style = "darkgrid")
plt.gcf().text(.2, .84, 'GENDER', fontsize = 20, fontweight = 'bold', color = 'Black', fontfamily = 'serif')
sns.lineplot(data = employment, x = 'Period', y = 'Unemployed', hue = 'Gender', style = 'Gender', dashes = False, palette = 'CMRmap', markers = ['o', '>'])
plt.show()
```
