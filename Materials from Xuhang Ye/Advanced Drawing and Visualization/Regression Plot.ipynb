```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import warnings

warnings.filterwarnings("ignore")
sns.set_style("whitegrid")

fish = pd.read_csv("fish.csv")
fish1 = fish[fish['Species'].isin(['Bream', 'Perch', 'Pike'])]

sns.regplot(x = fish1.Weight, y = fish1.Height, marker = '>', logx = True, line_kws = {'color': '#FF5722', 'alpha': 0.8, 'lw': 3}) # link_kws表示线条的属性，color表示颜色，alpha表示透明度，lw表示线条宽度
plt.show()
```

```python
# lmplot()函数
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import warnings

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid")

insurance = pd.read_csv('insurance.csv')
sns.lmplot(x = 'bmi', y = 'charges', data = insurance, aspect = 1.2, hue = 'smoker', height = 8, logx = True, markers = ['o', '>']) 
plt.show()
```
