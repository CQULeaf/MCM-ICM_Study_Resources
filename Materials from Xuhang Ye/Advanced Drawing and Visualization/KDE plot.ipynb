### 关于`kdeplot`函数中的`shade=True` 和 `shade_lowest=False` 参数

1. **shade = True**: 这个参数决定是否填充核密度估计（KDE）的等高线下方区域。`shade=True` 表示将填充这些区域，通常是使用`cmap`指定的颜色渐变。这有助于在视觉上突出密度较高的区域，使其更易于识别和理解。

2. **shade_lowest = False**: 这个参数用于控制是否填充密度最低的区域。以往的Seaborn版本默认会填充整个绘图区域的最低密度区域，但`shade_lowest=False`意味着不填充这些最低密度区域。这有助于集中视觉注意力在数据密集的区域，而不是在整个数据范围上，从而使得数据的重要部分更加突出。

综合这两个参数，该核密度估计图（KDE图）会更加集中于显示数据的重要部分，即高密度区域，而忽略掉较为稀疏的区域。这样的可视化有助于更好地理解和解释数据的分布特征。

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import warnings

warnings.filterwarnings("ignore")
sns.set_style("whitegrid")

insurance = pd.read_csv("insurance.csv")
sns.kdeplot(x = insurance.bmi, y = insurance.charges, cmap = "Reds", shade = True, shade_lowest = False) # 核密度
plt.show()
```

### 关于`swarmplot`的一些特殊参数

1. **palette**: `'coolwarm'`是一种颜色方案，提供了视觉上吸引人的颜色渐变，有助于区分不同的数据点。

2. **size**: 调整数据点的大小，使其更易于观察。这里设置为6（根据需要可以适当调整）。

3. **edgecolor**: 设置数据点边缘的颜色，这里使用`'gray'`为边缘添加灰色，有助于在数据点颜色较浅时增强可视化效果。

4. **linewidth**: 设置数据点边缘的线宽。这里设置为1，这样的设置既可以突出每个点，又不会过分强调边缘。

5. **alpha**: 设置数据点的透明度。`alpha=0.8`提供了一定的透明效果，可以帮助观察重叠的数据点，同时保持整体的清晰度。

通过这些参数，`swarmplot`不仅在视觉上更加吸引人，还能更有效地传达数据的重要特征。美观和可读性的平衡是关键，适当的参数调整可以根据数据的特性和展示目的来进行。

```python
sns.swarmplot(x = insurance.smoker, y = insurance.charges, palette = 'coolwarm', size = 6, edgecolor = 'gray', linewidth = 1, alpha = 0.8) # 蜂群图
plt.show()
```
