```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import warnings

warnings.filterwarnings("ignore")
sns.set_style("whitegrid")

fish = pd.read_csv("fish.csv")
print(fish.head())
```

```python
fish1 = fish[fish['Species'].isin(['Bream', 'Roach', 'Pike'])] #  筛选出那些 `Species` 列中属于 'Bream'、'Roach' 或 'Pike' 的行
print(fish1.head())
```

```python
sns.pairplot(fish1, hue='Species', palette = 'husl', size = 2)
plt.show()
```

```python
sns.pairplot(fish1, hue='Species', vars = ['Length1', 'Height', 'Width', 'Weight'], height = 3, palette = 'husl', aspect = 1)
plt.show()
```
